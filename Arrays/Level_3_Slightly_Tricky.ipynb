#1. Find the Largest Subarray Sum (Brute Force)

```python
arr = [1, -2, 3, 4, -1]

max_sum = arr[0]

for i in range(len(arr)):
  for j in range(i, len(arr)):
    current_sum = 0
    for k in range(i, j + 1):
      current_sum += arr[k]
    max_sum = max(max_sum, current_sum)

print(max_sum)

# Explanation:

# Try all possible subarrays

# Calculate sum of each

# Store the maximum sum found
```

#2. Find Intersection of Two Arrays

```python
arr1 = [1, 2, 3, 4]
arr2 = [3, 4, 5, 6]

intersection = []

for num in arr1:
  if num in arr2 and num not in intersection:
    intersection.append(num)

print(intersection)

# Explanation:

# Element must be in both arrays

# Avoid duplicates in result
```

#3. Find Union of Two Arrays

```python
arr1 = [1,2,3]
arr2 = [3,4,5]

union = []

for num in arr1:
  if num not in union:
    union.append(num)

for num in arr2:
  if num not in union:
    union.append(num)

print(union)

# Explanation:

# Add all unique elements from both arrays

# No duplicates allowed
```

#4. Check if Two Arrays Are Equal

```python
arr1 = [1,2,3]
arr2 = [1,2,3]

if len(arr1) != len(arr2):
  print(False)
else:
  equal = True
  for i in range(len(arr1)):
    if arr1[i] != arr2[i]:
      equal = False
      break
  print(equal)

# Explanation:

# Same length required

# Compare elements one by one
```

#5. Find Leaders in an Array

```python
arr = [16, 17, 4, 3, 5, 2]

leaders = []

for i in range(len(arr)):
  is_leader = True
  for j in range(i + 1, len(arr)):
    if arr[j] > arr[i]:
      is_leader = False
      break
    if is_leader:
      leaders.append(arr[i])

print(leaders)

# Explanation:

# Leader = greater than all elements to its right

# Check each element using nested loop
```

#6. Rearrange Array: Positive Then Negative

```python
arr = [-1,3,-2,4,-5,6]

result = []

for num in arr:
  if num >= 0:
    result.append(num)

for num in arr:
  if num < 0:
    result.append(num)

print(result)

# Explanation:

# First add all positive numbers

# Then add all negative numbers

# Order is preserved
```
